```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
import cv2
import time
from skimage import morphology
import seaborn as sns
plt.rcParams['figure.figsize'] = (20,12)
%config InlineBackend.figure_format = 'retina'
```

```python
def main(file_name):
    threshold_L = 15
    threshold_P = 0
    threshold_M = 10
    Lyso = cv2.imread(file_name + r"C1.tif",cv2.IMREAD_ANYDEPTH)
    ret_L ,th_L = cv2.threshold(Lyso,threshold_L,65535,cv2.THRESH_BINARY)
    th_L = (th_L).astype('uint8')
    th_L = th_L/255

    PDZD8 =  cv2.imread(file_name + "C2.tif",cv2.IMREAD_ANYDEPTH)
    ret_P ,th_P = cv2.threshold(PDZD8,threshold_P,65535,cv2.THRESH_BINARY)
    th_P = (th_P).astype('uint8')
    th_P = th_P/255
    PDZD8 = PDZD8 - threshold_P
    PDZD8 = PDZD8 * th_P

    Mito =  cv2.imread(file_name + "C3.tif",cv2.IMREAD_ANYDEPTH)
    ret_M ,th_M = cv2.threshold(Mito,threshold_M,65535,cv2.THRESH_BINARY)           
    th_M = (th_M).astype('uint8')
    th_M = th_M/255

    
    M = th_M - (th_M * th_L)
    L = th_L - (th_M * th_L)
    ML = th_M * th_L
    Other = np.ones_like(M) - (M + L + ML)
    
    
    PDZD8_alone = PDZD8 * Other
    PDZD8_with_M = PDZD8 * M
    PDZD8_with_L = PDZD8 * L
    PDZD8_with_ML = PDZD8 * ML

    return [np.sum(PDZD8_alone)/np.sum(PDZD8), np.sum(PDZD8_with_M)/np.sum(PDZD8), np.sum(PDZD8_with_L)/np.sum(PDZD8), np.sum(PDZD8_with_ML)/np.sum(PDZD8)]

```

```python
df = pd.DataFrame(data=None, index=None, columns= ["PDZD8_alone","PDZD8_on_only_Tomm20","PDZD8_on_only_Lamp1","PDZD8_on_Tomm20_and_Lamp1"], dtype=None, copy=False)
for i in range(1,10):
    file_name = r"/Volumes/Nakamura/Exp245/tif//KI_" + str(i).zfill(2) + r" - Deconvolved, Type Automatic.nd2_Z2_"
    df.loc[file_name] = main(file_name)
```
